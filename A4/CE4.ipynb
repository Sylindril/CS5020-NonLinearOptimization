```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def rosenbrock(x,y):
    '''Evaluates the Rosenbrock function at (x,y)'''
    return 100*np.power((y - np.power(x,2)),2) + np.power((x - 1),2)

def grad_X(x,y):
    '''Evaluates the gradient in the X-axis of the Rosenbrock function at (x,y)'''
    return -400*x*(y-np.power(x,2)) + 2*(x-1)

def grad_Y(x,y):
    '''Evaluates the gradient in the Y-axis of the Rosenbrock function at (x,y)'''
    return 200*(y-np.power(x,2))
```

### Gradient Descent

```python
#Setting up the stage for plotting
x = y = np.linspace(-5,5,500)
X,Y = np.meshgrid(x,y)
Z = rosenbrock(X,Y)
```

```python
#Defining the gradient step
x_iterate = 2
y_iterate = 2
alpha = 0.5

#Defining the convergence parameters
f_min = 0
argmin_X = argmin_Y = 0
```

```python
#Performing gradient descent
def gradient_step(x1, y1):
    xt,yt = x1, y1
    return (xt - alpha*grad_X(xt,yt), yt - alpha*grad_Y(xt,yt))

```

```python
#Plotting convergence
for i in range(10):
    
    x_iterate, y_iterate = gradient_step(x_iterate, y_iterate)
    errf = abs(round(rosenbrock(x_iterate,y_iterate),6) - 0)
    plt.figure(1)
    plt.plot(i, errf, marker="o", markersize=10, markeredgecolor="red",markerfacecolor="green")

   
```

```python
#plotting the function and points on it
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
```
